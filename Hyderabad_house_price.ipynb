```python
df['Location']= df['Location'].apply(lambda x: x.strip())
Location_count=df['Location'].value_counts()
Location_count
```

```python
Location_count_less_20 = Location_count[Location_count<= 20]
Location_count_less_20
```

```python
for i in Location_count_less_20:
    df['Location_'+ str(i) ]=np.where(df['Location']==str(i),1,0)
df.shape
```

```python
df['Location']=df['Location'].apply(lambda x: 'other' if x in Location_count_less_20 else x)
```

ML MODEL TRAINING SECTION

```python
from sklearn.model_selection import train_test_split

from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeRegressor


from sklearn.metrics import r2_score
```

ONE HOT ENCODE FOR CATEGORICAL VARIABLE

```python
df1=pd.get_dummies(df,drop_first=True,columns=['Location'])
df1.head()
```

```python
x=df1.drop(columns=['Price'])
y=df1['Price']
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=0)
print(x_train.shape)
print(x_test.shape)
```

SCALING THE FEATURES

```python
from sklearn.preprocessing import StandardScaler
scalar=StandardScaler()
scalar.fit(x_train)
x_train=scalar.transform(x_train)
x_test=scalar.transform(x_test)
```

LINEAR REGRESSION

```python
lr=LinearRegression()
lr.fit(x_train,y_train)
pred1=lr.predict(x_test)
r2_score(y_test,pred1)
```

```python
sns.distplot((y_test,pred1),bins=50)
```

DECISION TREE

```python
tree=DecisionTreeRegressor()
tree.fit(x_train,y_train)
pred3=tree.predict(x_test)
r2_score(y_test,pred3)
```

RANDOM FOREST

```python
 from sklearn.ensemble import RandomForestRegressor
 rf= RandomForestRegressor(max_depth=2, random_state=0)
 rf.fit(x_train,y_train)
 pred4=rf.predict(x_test)
 r2_score(y_test,pred4)
```
